```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 정의
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

# 초기 데이터 시각화
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()

# 가중치 초기값
w1 = 0
w2 = 0
b = 0
lr = 0.01
epochs = 2001
n = len(x1)

# 경사하강법
for i in range(epochs):
    y_hat = w1 * x1 + w2 * x2 + b
    error = y - y_hat

    dw1 = (2 / n) * sum(-x1 * error)
    dw2 = (2 / n) * sum(-x2 * error)
    db = (2 / n) * sum(-error)

    w1 = w1 - lr * dw1
    w2 = w2 - lr * dw2
    b = b - lr * db

    if i % 100 == 0:
        print(f'epoch = {i}, w1 = {w1:.4f}, w2 = {w2:.4f}, b = {b:.4f}')
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter3D(x1, x2, y)
        ax.plot3D(x1, x2, y_hat, 'r')  # 예측 선 그리기
        plt.show()

# 최종 예측 출력
y_hat = w1 * x1 + w2 * x2 + b
print(f'실제 점수: {y}\n예측 점수: {y_hat.round(2)}')
```
