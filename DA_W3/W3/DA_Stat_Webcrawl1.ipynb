```python
from bs4 import BeautifulSoup

import urllib.request
import urllib.parse

# url = 'https://www.example.com/search'
# url = 'https://www.google.com/search'
# url = 'https://www.google.com'
# url = 'https://www.chosun.co.kr/search'
# url = 'https://search.naver.com/search.naver'
url = 'https://kin.naver.com/search/list.naver'
# data = {'q': 'python'}
# data = {'q': '북한'}
# data = {'q': 'kpop'}
# data = {'query': 'kpop'} # naver 검색
data = {'query': '파이썬'} # naver 검색

data = urllib.parse.urlencode(data, encoding='utf-8')
# data = urllib.parse.urlencode(data)
url = url + '?' + data
# url = url

print(url)

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
    # 'Referer': 'https://www.example.com'
    # 'Referer': 'https://www.google.com'
    # 'Referer': 'https://www.chosun.co.kr'
    # 'Referer': 'https://search.naver.com'
    'Referer': 'https://kin.naver.com'
}

req = urllib.request.Request(url, headers=headers)
response = urllib.request.urlopen(req)

html = response.read()

print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

#### HTML 내의 태그 위치를 카피해서 가져오는 방법
- Chrome browser에서 네이버 지식인 서비스 사이트 접속 (검색어=파이썬): https://kin.naver.com/search/list.naver?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC
- 브라우저에서 f12 개발자툴 화면 띄워 찾고자 하는 위치 파악 (개발자 화면 상태에서 "Elements" 메뉴로 진입)
- 찾고자 하는 내용의 위에 마우스 오른 버튼 클릭: Copy > Copy Selector로 위치정보 카피
- 아래 위치정보 카피하여 soup.select() 호출

```python
li_list = soup.select("#suicidalPreventionBanner > div > ul > li.counsel_list.v1 > ul.counsel_info_list>li") # Copy-Selector 로 ul 태그의 'class' 속성값이 'counsel_info_list'인 블록 안의 'li' 태그 블록을 모두 추출

print("count: ", len(li_list))

for i in range(0, len(li_list)):
    print(li_list[i])
```

#### 네이버 지식인 답변 결과 리스트만 가져오기

- Chrome browser에서 파이썬, 다이아몬드 텍스트가 나오는 위치 찾기
- 위치정보를 Copy Selector로 카피
- soup.select()를 사용하여 적절하게 호출
- get_text()를 사용하여 결과 리스트 값에서 텍스트 정보만 출력

```python
li_list = soup.select("#s_content > div.section > ul.basic1>li") # Copy-Selector 로 ul 태그의 'class' 속성값이 'basic1'인 블록 안의 'li' 태그 블록을 모두 추출

print("count: ", len(li_list))

for i in range(0, len(li_list)):
    print(li_list[i].get_text())
```
