```python
import numpy as np
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([1.0, 1.0])
    b = -1.5
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1
    
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([1.0, 1.0])
    b = -0.5
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1
    
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([1.0, 1.0])
    b = -1.5
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 1
    else:
        return 0
    
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)

    return y

inputs = [(0,0), (0,1), (1,0), (1,1)]

print('\n[AND]')
for x1, x2 in inputs:
    y = AND(x1, x2)
    print(f'({x1}, {x2}) -> {y}')

print('\n[OR]')
for x1, x2 in inputs:
    y = OR(x1, x2)
    print(f'({x1}, {x2}) -> {y}')

print('\n[NAND]')
for x1, x2 in inputs:
    y = NAND(x1, x2)
    print(f'({x1}, {x2}) -> {y}')

print('\n[XOR]')
for x1, x2 in inputs:
    y = XOR(x1, x2)
    print(f'({x1}, {x2}) -> {y}')
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    return np.array(x>0, dtype = int)
def relu(x):
    return np.maximum(0,x)
def sigmoid(x):
    return 1/(1+np.exp(-x))
def dsigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))

x =np.arange(-10.0,10.0,0.01)
y = step(x)
z = relu(x)
plt.plot(x,y)
plt.plot(x,z,'r')
plt.xlim(-2.0, 2.0)
plt.ylim(-1.0, 3.0)
plt.yticks([0.0, 0.5, 1.0, 2.0])
plt.grid(axis='y')
plt.grid(axis='x')
plt.xlabel('x', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.show()

x = np.arange(-10.0,10.0,0.01)
y = sigmoid(x)
z = dsigmoid(x)
plt.plot(x,y)
plt.plot(x,z,'r')
plt.xlim(-5.0, 5.0)
plt.ylim(-0.1, 1.1)
plt.yticks([0.0, 0.5, 1.0])
plt.grid(axis='y')
plt.grid(axis='x')
plt.xlabel('x', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.show()
```
