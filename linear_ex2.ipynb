```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

# 하이퍼파라미터 초기화
w1, w2, b = 0.0, 0.0, 0.0
lr = 0.01
epochs = 2001
n = len(x1)

# 학습 전 데이터 시각화
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y, c='blue', label='실제 점수')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.legend()
plt.show()

# Gradient Descent 수행
for i in range(epochs):
    y_hat = w1 * x1 + w2 * x2 + b
    error = y - y_hat

    dw1 = -(2/n) * sum(x1 * error)
    dw2 = -(2/n) * sum(x2 * error)
    db = -(2/n) * sum(error)

    w1 -= lr * dw1
    w2 -= lr * dw2
    b -= lr * db

    # 중간 결과 시각화
    if i % 100 == 0:
        print(f'Epoch={i}, w1={w1:.4f}, w2={w2:.4f}, b={b:.4f}')
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter3D(x1, x2, y, c='blue', label='실제 점수')
        ax.plot3D(x1, x2, y_hat, 'r-', label='예측 점수')
        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.set_zlabel('Y')
        plt.legend()
        plt.show()

# 최종 예측 결과 출력
print(f'최종 실제 점수: {y}\n최종 예측 점수: {y_hat}')
```
