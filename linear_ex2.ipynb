```python
import numpy as np
import matplotlib.pyplot as plt

#  
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

#  
w1, w2, b = 0.0, 0.0, 0.0
lr = 0.01
epochs = 2001
n = len(x1)

#    
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y, c='blue', label=' ')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.legend()
plt.show()

# Gradient Descent 
for i in range(epochs):
    y_hat = w1 * x1 + w2 * x2 + b
    error = y - y_hat

    dw1 = -(2/n) * sum(x1 * error)
    dw2 = -(2/n) * sum(x2 * error)
    db = -(2/n) * sum(error)

    w1 -= lr * dw1
    w2 -= lr * dw2
    b -= lr * db

    #   
    if i % 100 == 0:
        print(f'Epoch={i}, w1={w1:.4f}, w2={w2:.4f}, b={b:.4f}')
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter3D(x1, x2, y, c='blue', label=' ')
        ax.plot3D(x1, x2, y_hat, 'r-', label=' ')
        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.set_zlabel('Y')
        plt.legend()
        plt.show()

#    
print(f'  : {y}\n  : {y_hat}')
```
