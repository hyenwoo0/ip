# 회귀 (Regression)

## 1. 선형 회귀 (Linear Regression)

- 1개의 독립변수를 사용해 종속변수 예측
- 예) 시험점수 = f(공부시간)

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()

w1 = 0
w2 = 0
b = 0

lr = 0.01
epochs = 2001
n = len(x1)

for i in range(epochs):
    y_hat = w1 * x1 + w2 * x2 + b
    # error = y - y_hat

    # dw1 = (2/n) * sum(-x1 * error)
    # dw2 = (2/n) * sum(-x2 * error)
    # db = (2/n) * sum(-error)

    error = y_hat - y

    dw1 = (2/n) * sum(x1 * error)
    dw2 = (2/n) * sum(x2 * error)
    db = (2/n) * sum(error)

    w1 = w1 - lr * dw1
    w2 = w2 - lr * dw2
    b = b - lr * db

    if i % 100 == 0:
        print(f'epoch = {i}, w1 = {w1:.4f}, w2 = {w2:.4f}, b = {b:.4f}, error = {error}')

        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.plot3D(x1, x2, y_hat, 'r')
        plt.show()

print(f'실제 점수: {y}\n예측 점수: {y_hat}')
```

## 2. 다중 선형 회귀 (Multiple Linear Regression)

- 2개 이상의 독립변수 사용해 종속변수 예측

- 예) 시험점수 = f(공부시간, 수면시간, 식사횟수)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

plt.scatter(x, y)
plt.show()

W = 0
b = 0
lr = 0.03
epochs = 2001
n = len(x)

for i in range(epochs):
    y_hat = W * x + b
    # error = y - y_hat

    # dW = (2/n) * sum(-x * error)
    # db = (2/n) * sum(-error)

    error = y_hat - y

    dW = (2/n) * sum(x * error)
    db = (2/n) * sum(error)

    W = W - lr * dW
    b = b - lr * db

    if i % 100 == 0:
        print(f'epoch = {i}, W = {W:.4f}, b = {b:.4f}, error = {error}')
        plt.scatter(x, y)
        plt.plot(x, y_hat, 'r')
        plt.show()

print(f'실제 점수: {y}\n예측 점수: {y_hat}')
```
